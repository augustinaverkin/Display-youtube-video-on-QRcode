```python
# 📦 STEP 1: Install GCS client library
!pip install --upgrade google-cloud-storage

# 📁 STEP 2: Upload key.json securely (WILL BE DELETED after use)
from google.colab import files
uploaded = files.upload()  # <- Select your key.json

key_file = next(iter(uploaded))  # Get uploaded filename
print(f"🔐 Using service account file: {key_file}")

# 🔐 STEP 3: Set environment and initialize GCS
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = key_file

from google.cloud import storage

# 🎯 CONFIGURATION – CHANGE THESE:
bucket_name = "imones-bucket-1"           # <-- change to your GCS bucket name
fotos_zip_blob_path = "fotos.zip" # <-- path to fotos.zip in bucket
local_zip_path = "fotos.zip"

# 🔽 STEP 4: Download fotos.zip from GCS
client = storage.Client()
bucket = client.bucket(bucket_name)
blob = bucket.blob(fotos_zip_blob_path)
blob.download_to_filename(local_zip_path)

print(f"✅ Downloaded {fotos_zip_blob_path} to {local_zip_path}")

# 🔓 STEP 5: Delete key.json immediately (for BDAR compliance)
os.remove(key_file)
print(f"🧹 Deleted service account file: {key_file}")

# 📦 STEP 6: Unzip fotos.zip (it contains images + annotations)
import zipfile

extract_dir = "fotos_data"
os.makedirs(extract_dir, exist_ok=True)

with zipfile.ZipFile(local_zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"✅ Extracted fotos.zip to: {extract_dir}")

# Optional: List contents
import os
print("\n📂 Unzipped contents:")
for root, dirs, files in os.walk(extract_dir):
    for name in files:
        print(f"📄 {os.path.join(root, name)}")
```
